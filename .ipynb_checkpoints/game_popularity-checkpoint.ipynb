```python
import numpy as np
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
```

```python
output_notebook()
```

```python
from bokeh.models import ColumnDataSource, HoverTool
```

```python
game_data = pd.read_csv('vgsales.csv', error_bad_lines=False)
```

```python
from bokeh.transform import factor_cmap
```

```python
game_data.Year = game_data.Year.astype(str)
game_data.Genre = game_data.Genre.astype(str)
```

```python
group = game_data.groupby(by=['Genre', 'Year'])
source = ColumnDataSource(group)
group.head(2)
```

```python
p = figure(plot_width=800, plot_height=300, title="Most popular type of genre per year", x_range=group, toolbar_location=None, tools='')
```

```python
p.xgrid.grid_line_color=None
p.xaxis.axis_label="Genres grouped by years they were most popular in"
p.xaxis.major_label_orientation=1.2
```

```python
index_cmap = factor_cmap('Genre_Year', palette=['#2b83ba', '#abdda4', '#2b83ba', '#abdda4', '#2b83ba', '#abdda4', '#2b83ba', '#abdda4', '#2b83ba', '#abdda4', '#2b83ba', '#abdda4'], factors=sorted(game_data.Genre.unique()), end=1)

p.vbar(x='Genre_Year', top='Year', width=1, source=source, line_color='white', fill_color=index_cmap, hover_line_color='darkgrey', hover_fill_color=index_cmap)
```

```python
show(p)
```
